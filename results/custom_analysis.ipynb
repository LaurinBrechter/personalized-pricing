```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plot_utils import plot_event_history
```

```python
sns.set_style("whitegrid")
```

```python
custom_log = pd.read_csv("custom_log.csv").replace({
    "run_id": {
        0: "low",
        1: "medium",
        2: "high"
    }
})
price_matrix = pd.read_csv("price_matrix.csv")
event_history = pd.read_csv("event_history.csv").dropna().replace({
    "run_id": {
        0: "low",
        1: "medium",
        2: "high"
    }
})
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))


sns.boxplot(custom_log, x="run_id", y="revenue", ax=ax[0])
event_history\
    .groupby(['run_id', 'event'])\
    .size()\
    .unstack()\
    .plot(kind='bar', ax=ax[1])
ax[0].set_title("Revenue per Strategy")
ax[0].set_xlabel("Strategy")
ax[1].set_title("Event Count")
ax[1].set_xlabel("Strategy")
```
